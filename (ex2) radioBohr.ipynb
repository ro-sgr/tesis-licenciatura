```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
nmin, nmax = 1, 5.1
n = np.arange(nmin, nmax, 1)
r = n**2
E = -1/(n**2)

n2 = np.arange(nmin, nmax, 0.1)
r2 = n2**2 
E2 = -1/(n2**2)

Rticklabels = []
Eticklabels = []
for i in n:
    Rticklabels.append(rf"$r_{{{int(i)}}}$")
    Eticklabels.append(rf"$E_{{{int(i)}}}$")
Eticklabels[3] = ''
```

```python
fig, axs = plt.subplots(1, 2, figsize=(15, 5))

axs[0].plot(n2, r2, linestyle='--', linewidth=1, marker=None, markersize=0, color='k')
axs[0].plot(n, r, linestyle='None', linewidth=None, marker='o', markersize=7, color='blue',
            label=r'$r_n = 4 \pi \varepsilon_0 \frac{n^2 \hbar^2}{Z m_e \, e^2}$')
axs[0].set_ylabel(r'Radio de Bohr ($r_n$)', fontsize='xx-large', labelpad=10)
axs[0].tick_params(axis='both', which='major', labelsize='xx-large')
axs[0].set_xlabel(r'Nmero cuntico principal ($n$)', fontsize='xx-large', labelpad=10)
# vals = np.array([-1])
axs[0].set_yticks(r)
axs[0].set_yticklabels(Rticklabels)
axs[0].set_ylim(-1, max(r)+1)
axs[0].legend(fontsize='xx-large', loc='upper left')
axs[0].axhline(0, linestyle=':', color='blue')
axs[0].grid()

axs[1].plot(n2, E2, linestyle='--', linewidth=1, marker=None, markersize=0, color='k', label=None)
axs[1].plot(n, E, linestyle='None', linewidth=None, marker='o', markersize=7, color='red',
            label=r'$E_n = - \frac{Z^2}{n^2} \left( \frac{m_e \, e^4}{32 \pi^2 \varepsilon_0 \hbar^2} \right)$')
axs[1].set_ylabel(r'Energa ($E_n$)', fontsize='xx-large', labelpad=10)
axs[1].tick_params(axis='both', which='major', labelsize='xx-large')
axs[1].set_xlabel(r'Nmero cuntico principal ($n$)', fontsize='xx-large', labelpad=10)
# vals = np.array([0.05])
axs[1].set_yticks(E)
axs[1].set_ylim(min(E)-0.05, 0.04)
axs[1].set_yticklabels(Eticklabels)
axs[1].legend(fontsize='xx-large', loc='lower right')
axs[1].axhline(0, linestyle=':', color='red')
axs[1].grid()

fig.subplots_adjust(wspace=0.25)
plt.show()
```
