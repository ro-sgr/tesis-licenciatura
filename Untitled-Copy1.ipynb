```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from elementos_matriz import *
```

```python
def elementos_matriz(d, a, RA, RB, ZA, ZB):
    """ Elementos de matriz para H2 con la base STO-3G

    Parámetros:
        d: vector de coeficientes de contracción
        a: vector de exponentes orbitales gaussianos
        (RA, RB): coord. del núcleo (A, B)
        (ZA, ZB): carga del núcleo (A, B)
    """
    # elementos de un cuerpo
    f11 = fpp(1, d, a, RA, RB, ZA, ZB)
    f33 = fpp(3, d, a, RA, RB, ZA, ZB)
    # elementos de dos cuerpos
    g1212 = gpqrs([1,2,1,2], d, a, RA, RB)
    g3434 = gpqrs([3,4,3,4], d, a, RA, RB)
    g1313 = gpqrs([1,3,1,3], d, a, RA, RB)
    g1331 = gpqrs([1,3,3,1], d, a, RA, RB)

    return f11, f33, g1212, g3434, g1313, g1331
```

```python
f11, f33, g1212, g3434, g1313, g1331 = [], [], [], [], [], []
```

```python
distances = np.arange(0.35, 7.66, 0.1) # de 0.35ua hasta 7.66ua en pasos de 0.1ua

for i in distances:
    RB = np.array([i, 0, 0])
    l1, l2, l3, l4, l5, l6 = elementos_matriz(d, a, RA, RB, ZA, ZB)
    f11.append(l1)
    f33.append(l2)
    g1212.append(l3)
    g3434.append(l4)
    g1313.append(l5)
    g1331.append(l6)
```

```python
plt.plot(distances, f11, label='f11')
plt.plot(distances, f33, label='f33')
plt.plot(distances, g1212, label='g1212')
plt.plot(distances, g3434, label='g3434')
plt.plot(distances, g1313, label='g1313')
plt.plot(distances, g1331, label='g1331')
plt.legend()
plt.grid()
plt.show()
```

```python
Gdist = [] # distintas matrices G según la distancia

for i in range(len(distances)):
    G = dict()
    for j in range(1,5):
        for k in range(1,5):
            # celda Gij
            G[f'{j}{k}'] = np.zeros((4,4))

    # g1212
    G['12'][0][1] = G['21'][1][0] = g1212[i]
    # g3434
    G['34'][2][3] = G['43'][3][2] = g3434[i]
    # g1313
    G['13'][0][2] = G['14'][0][3] = G['23'][1][2] = G['24'][1][3] = g1313[i]
    G['31'][2][0] = G['41'][3][0] = G['32'][2][1] = G['42'][3][1] = g1313[i]
    # g1331
    G['12'][2][3] = G['13'][2][0] = G['14'][2][1] = g1331[i]
    G['21'][3][2] = G['23'][3][0] = G['24'][3][1] = g1331[i]
    G['31'][0][2] = G['32'][0][3] = G['34'][0][1] = g1331[i]
    G['41'][1][2] = G['42'][1][3] = G['43'][1][0] = g1331[i]

    # guardar el valor de G
    Gdist.append(G)
```

```python
fig, ax = plt.subplots()

for dist in range(len(distances)):
    Gblock = np.block([[Gdist[dist]['11'], Gdist[dist]['12'], Gdist[dist]['13'], Gdist[dist]['14']],
          [Gdist[dist]['21'], Gdist[dist]['22'], Gdist[dist]['23'], Gdist[dist]['24']],
          [Gdist[dist]['31'], Gdist[dist]['32'], Gdist[dist]['33'], Gdist[dist]['34']],
          [Gdist[dist]['41'], Gdist[dist]['42'], Gdist[dist]['43'], Gdist[dist]['44']]])
    
    ax.matshow(Gblock, cmap=plt.cm.Blues)
    
    nombre = f'G/{dist}.png'
    plt.savefig(nombre, dpi=50, format='png')
```

```python
def energy_HF(d, a, RA, RB, ZA, ZB):
    """ Cálculo de la energía mediante el método de Hartree-Fock (HF)
    
     Parámetros:
        d: vector de coeficientes de contracción
        a: vector de exponentes orbitales gaussianos
        (RA, RB): coord. del núcleo (A, B)
        (ZA, ZB): carga del núcleo (A, B)
    """
    f11, f33, g1212, g3434, g1313, g1331 = elementos_matriz(d, a, RA, RB, ZA, ZB)
    h_nuc = 1/(np.linalg.norm(RA-RB))
    E0 = 2*f11 + g1212
    print('f11: ', f11)
    # print('f33: ', f33)
    print('g1212: ', g1212)
    # print('g3434: ', g3434)
    # print('g1313: ', g1313)
    # print('g1331: ', g1331)
    Etot = E0 + h_nuc
    
    return Etot
```
