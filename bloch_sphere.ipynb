```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_bloch_multivector
```

### Hidrógeno

```python
qc = QuantumCircuit(1)
qc.id(0)
qc.z(0)

qc.draw()
```

```python
state = Statevector(qc)
plot_bloch_multivector(state)
```

### Helio

```python
qc = QuantumCircuit(2)
qc.id(0)
qc.z(0)
qc.barrier()
#########
qc.id(1)
qc.z(1)
qc.barrier()
#########
qc.id(0)
qc.id(1)
#
qc.id(0)
qc.z(1)
#
qc.z(0)
qc.id(1)
#
qc.z(0)
qc.z(1)

qc.draw()
```

```python
state = Statevector(qc)
plot_bloch_multivector(state)
```

### Molécula de hidrógeno ionizada

```python
qc = QuantumCircuit(2)
qc.id(0)
qc.z(0)
qc.barrier()
#########
qc.id(1)
qc.z(1)
qc.barrier()
#########

qc.draw()
```
