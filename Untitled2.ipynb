```python
import numpy as np
import matplotlib.pyplot as plt
import os
from matplotlib.colors import Normalize

# Example data: 4 timesteps of a symmetric 3x3 matrix
G_all = [
    np.array([[1, 2, 3],  # t=0
              [2, 4, 5],
              [3, 5, 1]]),
    np.array([[2, 3, 4],  # t=1
              [3, 6, 7],
              [4, 7, 2]]),
    np.array([[3, 4, 5],  # t=2
              [4, 8, 9],
              [5, 9, 3]]),
    np.array([[4, 5, 6],  # t=3
              [5, 10, 11],
              [6, 11, 4]])
]

# Create directory 'G' if it doesn't exist
os.makedirs('Gtest', exist_ok=True)

# Step 1: Define groups based on equality constraints
groups = {
    'Group1': [(0, 5), (5, 0)],      # g1212 and g2121
    'Group2': [(15, 10), (10, 15)],  # g3434 and g4343
    'Group3': [(0, 10), (0, 15), (5, 10), (5, 15), (10, 0), (15,0), (10,5), (15, 5)],  # g1313, g1414, g2323, g2424, g3131, g4141, g3232 and g4242
    'Group4': [(1, 1)],          # g1234, g1313, g1423, g2134, g2314, g
    'Group5': [(1, 2), (2, 1)]   # 
}

# Step 2: Compute global min/max for each group
group_stats = {}
for group_name, indices in groups.items():
    values = []
    for t, G in enumerate(G_all):
        for (i, j) in indices:
            values.append(G[i, j])
    if values:
        group_stats[group_name] = {
            'min': min(values),
            'max': max(values),
            'indices': indices
        }

# Step 3: Assign a colormap to each group
cmaps = ['Blues', 'Greens', 'Reds', 'Purples', 'Oranges']
group_cmaps = {group: plt.get_cmap(cmaps[i]) for i, group in enumerate(groups.keys())}

# Step 4: Plot and save each timestep
for t, G in enumerate(G_all):
    fig, ax = plt.subplots(figsize=(6, 6))
    colored_G = np.zeros((*G.shape, 4))  # RGBA array
    
    for group_name, stats in group_stats.items():
        cmap = group_cmaps[group_name]
        norm = Normalize(vmin=stats['min'], vmax=stats['max'])
        for (i, j) in stats['indices']:
            colored_G[i, j] = cmap(norm(G[i, j]))
    
    ax.imshow(colored_G, interpolation='nearest')
    ax.set_title(f"Time t = {t}", fontsize=12, weight='bold')
    ax.set_xticks([])
    ax.set_yticks([])
    
    # Save the figure
    plt.savefig(f"Gtest/timestep_{t}.png", bbox_inches='tight', dpi=150)
    plt.close()

print("All plots saved in the 'G' directory!")
```
